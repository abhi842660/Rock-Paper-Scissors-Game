```python
# Rock,Paper,Scissors Game
import tkinter as tk
from random import choice
from tkinter import messagebox

# Create the main application window
root = tk.Tk()
root.title("Rock, Paper, Scissors Game")
root.geometry("400x400")

# Game options
options = ["Rock", "Paper", "Scissors"]

# Initialize score
player_score = 0
computer_score = 0
tie_score = 0

# Function to decide the winner
def play(player_choice):
    global player_score, computer_score, tie_score
    computer_choice = choice(options)
    
    # Update labels for choices
    player_label.config(text=f"Player: {player_choice}")
    computer_label.config(text=f"Computer: {computer_choice}")
    
    # Determine the result
    if player_choice == computer_choice:
        result = "It's a tie!"
        tie_score += 1
    elif (player_choice == "Rock" and computer_choice == "Scissors") or \
         (player_choice == "Paper" and computer_choice == "Rock") or \
         (player_choice == "Scissors" and computer_choice == "Paper"):
        result = "You win!"
        player_score += 1
    else:
        result = "You lose!"
        computer_score += 1
    
    # Update result label and scores
    result_label.config(text=result)
    score_label.config(text=f"Score: Player {player_score} - Computer {computer_score} - Ties {tie_score}")

# Function to reset the game
def reset_game():
    global player_score, computer_score, tie_score
    player_score, computer_score, tie_score = 0, 0, 0
    result_label.config(text="")
    player_label.config(text="Player: ")
    computer_label.config(text="Computer: ")
    score_label.config(text="Score: Player 0 - Computer 0 - Ties 0")

# GUI Elements

# Labels for player and computer choices
player_label = tk.Label(root, text="Player: ", font=("Arial", 14))
player_label.pack(pady=10)

computer_label = tk.Label(root, text="Computer: ", font=("Arial", 14))
computer_label.pack(pady=10)

# Label for the result of the game
result_label = tk.Label(root, text="", font=("Arial", 16))
result_label.pack(pady=10)

# Label for displaying scores
score_label = tk.Label(root, text="Score: Player 0 - Computer 0 - Ties 0", font=("Arial", 14))
score_label.pack(pady=10)

# Buttons for Rock, Paper, Scissors
rock_button = tk.Button(root, text="Rock", width=15, command=lambda: play("Rock"))
rock_button.pack(pady=5)

paper_button = tk.Button(root, text="Paper", width=15, command=lambda: play("Paper"))
paper_button.pack(pady=5)

scissors_button = tk.Button(root, text="Scissors", width=15, command=lambda: play("Scissors"))
scissors_button.pack(pady=5)

# Button to reset the game
reset_button = tk.Button(root, text="Reset Game", width=15, command=reset_game)
reset_button.pack(pady=10)

# Run the Tkinter event loop
root.mainloop()
```
